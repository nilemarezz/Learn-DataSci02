## Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np;
```

```python
# ใช้ numpy สร้าง Array 11 ตัว เริ่มจาก 0 เพิ่มทีละ 0.5
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
# Functional Method
plt.plot(x,y)
plt.xlabel('X Lable')
plt.ylabel('Y Lable')
plt.title('title')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r') # <-- r คือ สี

plt.subplot(1,2,2)
plt.plot(y,x,'b')

```

```python
# Object Oriented Method
# สร้าง figure คือ กระดาษเปล่า
fig = plt.figure();
asxes = fig.add_axes([0.1,0.1 ,0.8,0.8])
asxes.set_xlabel('X Lable')
asxes.set_ylabel('Y Lable')
asxes.set_title('title')
asxes.plot(x,y)
```

```python
fig = plt.figure();
axes1 = fig.add_axes([0.1,0.1 ,0.8,0.8])
axes1.plot(x,y)
axes1.set_title('Large plot')
axes2 = fig.add_axes([0.3,0.5,0.4,0.3])
axes2.plot(y,x)
axes2.set_title('Small plot')
```

```python
# Subplot ด้วย OO 
#(แถวและคอลัมน์ อารม4เหลี่ยมเช่น 1*2 , 2*2 ก็จะมี 4)
fig,axes = plt.subplots(nrows = 1,ncols =2)
axes[0].plot(x,y)
axes[0].set_title('Index 0')
axes[1].plot(y,x)
axes[1].set_title('Index 1')
```

#### Figure Size and DPI

```python
# ลด size ของ กราฟ ด้วย figsize=(กว้าง,ยาว)
fig = plt.figure(figsize=(8,2))
axes3 = fig.add_axes([0.1,0.1,0.8,0.8])
axes3.plot(x,y,label = 'x Square')
axes3.plot(y,x,label = 'x Cube')
axes3.legend()
```

```python
fig2,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
# export as png (dpi คือ ขนาด)
fig2.savefig('my_picture.png',dpi=200)
```
