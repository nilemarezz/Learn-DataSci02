```python
import seaborn as sns
import pandas as pd
```

```python
%matplotlib inline
```

## Distplot

```python
tips = sns.load_dataset('tips')
```

```python
tips.head(5)
```

```python
sns.distplot(tips['total_bill'])
```

```python
# bins คือจำนวนกราฟ
sns.distplot(tips['total_bill'],bins = 100)
```

```python
# ใส่ kind = '' เพื่อเปล่ยนประเภทเช่น hex,reg,kde
sns.jointplot(x = 'total_bill' ,y= 'tip',data=tips)
```

```python
sns.pairplot(tips)
```

## Categorical plot

```python
import numpy as pd;
```

```python
# ดู mean x=ประเภท y=ค่าที่คำนวนได้
sns.barplot(x='sex',y='total_bill',data = tips)
```

```python
# นับจำนวน เปลี่ยนสีได้ที่่ pallete 
sns.countplot(x='day',data=tips,hue = 'smoker',palette = 'Set1')
# เช็คแต่ละแท่ง
tips[(tips['day'] == 'Thur' ) & (tips['smoker'] == 'No')].count()
```

```python
sns.boxplot(x='day',y='total_bill',data = tips,hue = 'smoker')
```

## Regression Plot

```python
sns.lmplot(x='total_bill',y= 'tip' ,data = tips ,hue = 'sex',palette = 'rainbow')
```
